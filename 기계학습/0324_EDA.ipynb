# group by

```python
from itertools import groupby
```

```python
a = [1,1,1,1,2,2,1,2,3,1,3,3,3]
```

```python
t = groupby(a)
#  [1, 2, 1, 2, 3, 1, 3]  
#        
#  [1,2,3]   
```

```python
next(t)
```

```python
t = groupby(a)
for i, j in t:
    print('++++++')
    for a in j:
        print('-----')
        print(a)
```

```python
from operator import itemgetter
```

```python
t = groupby(a, key=itemgetter)
for i, j in t:
    print(i, j)
```
