# n-gram

n-gram은 약간 최대우도를 이용하는 거다\
단어가 조건부로 주어졌을 때의 조건부 확률을 최대로 하는 단어를 찾는 느낌같음\
단어 간의 연관을 이런 식으로 분석함\
같이 자주 나오는 단어의 힘을 이용하는 것.\
그리고 그 조건부의 갯수에 따라 2-gram(바이그램), 3-gram(트라이그램) ,, 뭐 이런 식\
띄어쓰기 기준으로 건너뛰지 않고 1개씩 옆으로 이동하면서 수행된다.

```python
def ngram(t, n=2): # 2번째 방법(Padding 없이)
    tokens = t.split()
    gram = list()
    
    if len(tokens) < n: # 최소 토큰이 2개는 있어야 함
        return tokens
    for i in range(len(tokens)-(n-1)):
        gram.append(' '.join(tokens[i:i+n]))
    return gram
```

```python
ngram('유구한 역사와 전통에 빛나는 우리', n=3)
```

```python
ngram(data[-1])
```

```python
import re
from string import punctuation

p1 = re.compile('[{}]'.format(re.escape(punctuation)))
p2 = re.compile('\s+')

ngram(p2.sub(' ', p1.sub(' ', data[-1]))) # p1을 처리하고 그걸 p2에 넣고 이제 ngram을 돌려준다. 위에보다 더 좋음
```

```python
model = dict()

for gram in ngram(p2.sub(' ', p1.sub(' ', data[-1]))):
    if gram in model:
        model[gram] += 1
    else:
        model[gram] = 1
```
