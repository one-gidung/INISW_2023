```python
# 한소희 검색결과
url = 'https://www.google.co.kr/search?q=%ED%95%9C%EC%86%8C%ED%9D%AC&sxsrf=AJOqlzW-o71_Jf3L4_7VXTCBvi4KKFg5wA%3A1678417833888&ei=qZ8KZLzlNcX--Qan05yIBg&ved=0ahUKEwj83ZfWsdD9AhVFf94KHacpB2EQ4dUDCA8&uact=5&oq=%ED%95%9C%EC%86%8C%ED%9D%AC&gs_lcp=Cgxnd3Mtd2l6LXNlcnAQAzIECCMQJzIICC4QgAQQsQMyBQgAEIAEMgUIABCABDIECAAQAzIFCAAQgAQyBQgAEIAEMgUIABCABDIFCAAQgAQyBQguEIAEOhEILhCABBCxAxCDARDHARDRAzoLCAAQgAQQsQMQgwE6BwgjEOoCECc6DQguEIAEEBQQhwIQsQM6CwguEIAEELEDEIMBOgoIABCABBAUEIcCOhEILhCABBCxAxCDARDHARCvAToLCC4QgAQQxwEQrwE6CAgAEIAEELEDSgQIQRgAUABYjQpgmgtoAnABeAGAAaoBiAGvCJIBAzAuOZgBAKABAbABCsABAQ&sclient=gws-wiz-serp'
header = {
    'user-agent': 'mybot'
}
req = Request(url, headers=header, method='GET')
```
