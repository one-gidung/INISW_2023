그 당시 사용한게 sigmoid activation function이다.

**sigmoid**

장점
- step function과 가장 비슷한 활성함수로 시뮬레이션이 가능하다.

단점
- 학습이 느리다.
- 포화, saturated되는 문제점, 값이 다 0이 되버린다. gradient vanishing 문제가 생긴다.
- 그래서 layer를 많이 쌓을 수 없었다.
- 기울기가 소실이 되어 학습이 잘 되지 않아 깊은 모델 쌓을 수가 없다.
- saturated
