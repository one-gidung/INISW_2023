```python
basis = base.fillna(0).copy()
basis
```

```python
basis_corr = basis.T.corr().copy()
basis_corr
```

```python
# 나랑 유사도가 비슷한 사람을 추천
# 근데 여기서 비슷한 유저가 "안 본" 영화를 추천해줘야 하겠지?
basis_corr.loc[7].sort_values(ascending=False)[1:6]
```

```python
# 나랑 가까운 애들 찾는 함수 그냥 만들었어
def nearest(n, x=5):
    return basis_corr.loc[n].sort_values(ascending=False)[1:x+1]
```

```python
# 얼마나 편해
nearest(7,4)
```

```python
# 650번 애가 영화를 많이 봤네... 311개...
# 이걸 함수로 만들어보자
data[data['user id']==650]['item id']
```

```python
# 자 다시 평점 높은 거로 함수 수정해보장
temp = data[data['user id']==650]
temp
```

```python
temp[temp.rating>=4]
```

```python
def seen_movie(n, rank=4):
    temp = data[data['user id']==n]
    return set(temp[ temp.rating>=rank ]['item id'].values)
```

```python
# 서로 안 본 영화 찾아야겠지? 차집합해보자 이걸 이용해서
seen_movie(650), seen_movie(7)
```

```python
# set으로 집합 만들어서 빼주자 그럼 차집합
# 근데 여기서 평점이 좋은 애들을 추천해주고 싶지 않니?
# 함수를 다시 수정하자
# 함수 내에서 set으로 return을 수정해줬으니 여기서 set을 다시 지웠다.
seen_movie(650)-seen_movie(7)
```

- 다시 함수 수정하자

```python
# 나랑 가까운 애들 찾는 함수 그냥 만들었어
# index를 뽑아보자 

def nearest(n, x=5):
    return basis_corr.loc[n].sort_values(ascending=False)[1:x+1].index
```

```python
nearest(7)
```

```python
# 이거 아니면 for문으로 or로 많이 돌려야하는데
# 벡터라이즈 연산이 돼서 얼마나 멋진가..
# loop없이... 메모리 번지 던져서 받는다 아주 빠르다 동시에 CPU multicore까지 지원한다

data[data['user id'].isin(nearest(7))]
```

- 함수 다시 수정

```python
# 나하고 가까운애 기본 5명이 본 영화들
# 근데 중복되어 있을거야
# set을 써주자

# item
def nearest(n, x=5):
    ind = basis_corr.loc[n].sort_values(ascending=False)[1:x+1].index
    return set(data[data['user id'].isin(ind)]['item id'])
```

```python
nearest(7)
```
